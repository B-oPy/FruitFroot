```python
import pandas as pd 
import os 
import json
from pandas.io.json import json_normalize
json_list = os.listdir('./json/')
json_list
```

```python
i = './json/V006_80_1_01_01_01_23_1_1483y_20200928_71.jpg.json'
with open(i, 'r') as f:
        dict = json.loads(f.read())
        df1 = pd.json_normalize(dict['description']) 
        df2 = pd.json_normalize(dict['annotations']) 
        df3 = pd.concat([df1,df2], axis=1)
df3
```

```python
df = pd.DataFrame()
for i in json_list:
    with open('./json/'+i, 'r') as f:
        dict = json.loads(f.read())
        df1 = pd.json_normalize(dict['description']) 
        df2 = pd.json_normalize(dict['annotations']) 
        df3 = pd.DataFrame(pd.concat([df1,df2], axis=1))
        df = pd.concat([df,df3], axis=0)
df
```
